```python
data['occupation'] = data['occupation'].replace('?',np.nan)
data['workclass'] = data['workclass'].replace('?',np.nan)
data['native-country'] = data['native-country'].replace('?', np.nan)
```

```python
from sklearn.model_selection import train_test_split
X = data[['workclass','education','relationship','race','sex','hours-per-week','native-country']]
y= data['salary']
X_train, X_test, Y_train,Y_test = train_test_split(X,y, random_state = 20, test_size = 0.3)
```

```python
Y_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(handle_unknown = 'ignore')
encoder.fit(X_train)
encoder.fit(X_test)
X_train = encoder.transform(X_train)
X_test = encoder.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

model = []
names = ['Decisiontree','RandomForest','SVC']


model.append(DecisionTreeClassifier())
model.append(RandomForestClassifier())
model.append(SVC())


decisiontree = DecisionTreeClassifier()
Random = RandomForestClassifier()
svc = SVC()
print(model)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score

for i in range(0, len(model)):
    cv = cross_val_score(model[i],X_train,Y_train,cv = 10, scoring = 'accuracy')
    model[i].fit(X_train,Y_train)
    y_pred = model[i].predict(X_test)
    score = accuracy_score(Y_test,y_pred)
    print('{0} : {1}'.format(names[i],score))
```

### The Support vector classifier performed well with the dataset. 

```python
from sklearn import metrics

svc = SVC(C = 1, gamma= 'scale', kernel='rbf', verbose= 2)
svc.fit(X_train,Y_train)
ypred1 = svc.predict(X_test)
print('The accuracy score is ',metrics.accuracy_score(Y_test, ypred1))
print(metrics.classification_report(Y_test, ypred1))
print(metrics.confusion_matrix(Y_test, ypred1))
```
